```python
from matplotlib import pyplot as plt
import numpy as np
import SimpleITK as sitk
import os
```

```python
#Set path to dcm and nrrd files 
image_path = os.path.join("C:/Users/samkr/Downloads/DICOM FILES/annotated/9008820_RDY")
segmented_path = os.path.join(image_path, "NRRD", "9008820-label.nrrd")
segmentation = sitk.ReadImage(segmented_path)
```

```python
seg1 = sitk.GetArrayFromImage(segmentation)
for p in range(16):
    current_seg = seg1[p]
    new_seg = np.zeros([256,512,9])
    new_seg[0][0][8]

    for y,a in enumerate(current_seg):
        for x,b in enumerate(a):
            if b>0:
                new_seg[y][x][b-1]=1
    threshold, upper, lower = 0.5, 1, 0
    new_seg[new_seg>threshold] = upper
    new_seg[new_seg<=threshold] = lower

    for i in range (9):
        plt.imshow(new_seg[:,:,i])
        plt.show()
```
